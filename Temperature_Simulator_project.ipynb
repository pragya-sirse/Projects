```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def temperature_simulator(num_days,mean_temp,temp_variation=5):
  temperature_data=np.random.normal(loc=mean_temp,scale=temp_variation,size=num_days)
  return temperature_data
```

```python
num_days=365
mean_temperature=25
temp_variation=5
```

```python
temperature_data=temperature_simulator(num_days,mean_temperature,temp_variation)
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.plot(temperature_data, label='Simulated Temperature')
plt.axhline(y=mean_temperature,color='y',linestyle='--',label='Mean Temperature')
plt.title('Simulated Daily Temperature Over Time')
plt.xlabel('nums_days')
plt.ylabel('Temperature (c)')
plt.legend()
plt.show()
```
